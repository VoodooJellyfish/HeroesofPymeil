```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_df = pd.read_csv(file_to_load)
purchase_df.head()
```

## Player Count

* Display the total number of players


```python
total_players = purchase_df[["SN"]].nunique()
total_df = pd.DataFrame(total_players, columns = ['Total Players'])
total_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
items = int(purchase_df[["Item ID"]].nunique())
average_price = int(purchase_df[["Price"]].mean())
purchases = int(purchase_df[["Purchase ID"]].count())
revenue = int(purchase_df[["Price"]].sum())

data = {"Number of Unique Items": items, "Average Price": average_price, "Number of Purchases": purchases, "Total Revenue": revenue}
summary_df = pd.DataFrame(data, index= [0])
summary_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python

unique_users = []
male = 0
female = 0
other = 0

for column, row in purchase_df.iterrows():
    
    if row[1] not in unique_users:
        if row[3] == "Male":
            male = male + 1
        elif row[3] == "Female":
            female = female + 1
        else:
            other = other + 1
        unique_users.append(row[1])
    else:
        pass

total_players = int(total_players)
perMale = (male / total_players) * 100
perFemale = (female / total_players) * 100
perOther = (other / total_players) * 100

gender_data = {None:["Male", "Female", "Other"],"Total Count": [male, female, other], "Percentage of Users": [perMale, perFemale, perOther]}
gender_df = pd.DataFrame(gender_data)
gender_df.set_index(None)

```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
gender_group = purchase_df.groupby("Gender")
purchase_count = gender_group["SN"].count()
average_price = gender_group["Price"].mean()
purchase_total = gender_group["Price"].sum()

unique_gender_group = purchase_df.groupby("Gender")["SN"]
average_per_person = purchase_total / unique_gender_group.nunique() 

purchase_summary = pd.DataFrame({"Purchase Count": purchase_count, "Average Purchase Price": average_price,
                   "Total Purchase Value": purchase_total, "Avg Total Purchase per Person":average_per_person  })
purchase_summary
```

## Age Demographics



* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python

bins = [0, 9, 14, 19, 24, 29, 34, 39, 40]
labels = ["<10", "10-14", "15-19" ,"20-24", "25-29", "30-34","35-39", "40+"]
purchase_df["Age Group"] = pd.cut(purchase_df["Age"], bins, labels=labels)

age_grouped = purchase_df.groupby("Age Group")
age_count = age_grouped["SN"].nunique()
age_percent = (age_count / total_players) * 100
age_summary = pd.DataFrame({"Total Count": age_count, "Percentage of Players":age_percent})
age_summary
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
age_purchase_group = purchase_df.groupby("Age Group")
age_purchase_count = age_purchase_group["SN"].count()
age_average_price = age_purchase_group["Price"].mean()
age_purchase_total = age_purchase_group["Price"].sum()

unique_age_group = purchase_df.groupby("Age Group")["SN"]
age_average_per_person = age_purchase_total / unique_age_group.nunique()

age_purchase_summary = pd.DataFrame({"Purchase Count": age_purchase_count, "Average Purchase Price": age_average_price,
                   "Total Purchase Value": age_purchase_total, "Avg Total Purchase per Person":age_average_per_person })
age_purchase_summary
```

## Most Popular Items

```python
sn_group = purchase_df.groupby("SN")
sn_count = sn_group["Purchase ID"].count()
sn_average_price = sn_group["Price"].mean()
sn_total_purchase = sn_group["Price"].sum()
age_purchase_summary = pd.DataFrame({"Purchase Count": sn_count, "Average Purchase Price": sn_average_price,
                   "Total Purchase Value": sn_total_purchase})
age_purchase_summary.sort_values(by=["Purchase Count"], ascending=False).head()
```

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
item_df = purchase_df[["Item ID", "Item Name", "Price"]]

item_group = item_df.groupby(['Item ID', "Item Name"])
item_count = item_group["Price"].count()
item_average = item_group["Price"].mean()
item_total_value = item_count * item_average
item_purchase_summary = pd.DataFrame({"Purchase Count": item_count, "Average Purchase Price": item_average,
                   "Total Purchase Value": item_total_value})
item_purchase_summary.sort_values(by=["Purchase Count"], ascending=False).head()
```
